## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
df_gasolina_2000 = pd.read_csv('analise-precos-gasolina/gasolina_2000+.csv')
df_gasolina_2010 = pd.read_csv('analise-precos-gasolina/gasolina_2010+.csv')
df_gasolina = (
    pd.concat([df_gasolina_2000, df_gasolina_2010], ignore_index=True)
    .rename(columns={'Unnamed: 0': 'ID'})
    .set_index('ID')
)
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
df_gasolina['DATA INICIAL'] = pd.to_datetime(df_gasolina['DATA INICIAL'])
df_gasolina['DATA FINAL'] = pd.to_datetime(df_gasolina['DATA FINAL'])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
df_gasolina['DATA FINAL ANO E MES'] = df_gasolina['DATA FINAL'].dt.to_period('M')
df_gasolina[['DATA FINAL ANO E MES', 'DATA FINAL']]
```

### 6. Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

```python
df_gasolina.value_counts(['PRODUTO'])
```

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
df_gasolina_comum = df_gasolina[df_gasolina['PRODUTO'] == 'GASOLINA COMUM']
df_gasolina_comum
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
df_gasolina_comum.loc[df_gasolina_comum['DATA FINAL ANO E MES'] == '2008-08', 'PREÇO MÉDIO REVENDA'].mean() 
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
df_gasolina_comum.loc[df_gasolina_comum['DATA FINAL ANO E MES'] == '2014-05', ['PREÇO MÉDIO REVENDA']]
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
df_gasolina_comum.loc[
    df_gasolina_comum['PREÇO MÉDIO REVENDA'] >= 5,
    ['PREÇO MÉDIO REVENDA', 'DATA FINAL ANO E MES', 'ESTADO']
]
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
df_gasolina.loc[
    (df_gasolina['DATA FINAL ANO E MES'].dt.year == 2012) &
    (df_gasolina['REGIÃO'] == 'SUL'),
    ['REGIÃO', 'ESTADO', 'PREÇO MÉDIO REVENDA', 'DATA FINAL ANO E MES']
]
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
#df_gasolina_rj = df_gasolina.loc[df_gasolina['ESTADO'] == 'RIO DE JANEIRO']
#df_gasolina_rj.groupby(df_gasolina_rj['DATA FINAL ANO E MES'].dt.year)['PREÇO MÉDIO REVENDA'].mean()

df_gasolina.loc[df_gasolina['ESTADO'] == 'RIO DE JANEIRO'] \
    .groupby(df_gasolina.loc[df_gasolina['ESTADO'] == 'RIO DE JANEIRO', 'DATA FINAL ANO E MES'].dt.year)['PREÇO MÉDIO REVENDA'] \
    .mean()
```
